```python
import math
import svgwrite

def polar_to_cartesian(cx, cy, r, angle_deg):
    """
    ì¤‘ì‹¬(cx, cy)ì—ì„œ ë°˜ì§€ë¦„ r, ê°ë„ angle_deg(ë„ ë‹¨ìœ„) ìœ„ì¹˜ì— í•´ë‹¹í•˜ëŠ” (x, y) ì¢Œí‘œ ë°˜í™˜.
    0ë„ëŠ” í™”ë©´ ì˜¤ë¥¸ìª½(+xì¶•) ë°©í–¥, ì‹œê³„ ë°˜ëŒ€ ë°©í–¥ì´ ê°ë„ ì¦ê°€ ë°©í–¥.
    """
    rad = math.radians(angle_deg)
    x = cx + r * math.cos(rad)
    y = cy + r * math.sin(rad)
    return x, y

def create_circular_diagram(
    data,
    center_text="TEXT HERE",
    filename="diagram.svg",
    size=600,
    # ì¤‘ì•™ ì› ìŠ¤íƒ€ì¼
    center_circle_radius=60,
    center_circle_fill="#333333",
    center_circle_text_color="#FFFFFF",
    # ë°”ê¹¥ ì›(í…Œë‘ë¦¬) ìŠ¤íƒ€ì¼
    outer_ring_radius=180,            # ë°”ê¹¥ìª½ ì‘ì€ ì›ë“¤ì´ ìœ„ì¹˜í•  'ì›í˜• ê¶¤ë„' ë°˜ì§€ë¦„
    outer_ring_stroke="#CCCCCC",
    outer_ring_stroke_width=2,
    # ê° ì•„ì´í…œ ì›(ì œëª© ì›) ìŠ¤íƒ€ì¼
    item_circle_radius=30,
    item_circle_stroke="#333333",
    item_circle_stroke_width=2,
    item_circle_fill="#FFFFFF",
    # í°íŠ¸ ì„¤ì •
    title_font_size=14,
    content_font_size=12,
):
    """
    data: [
      {
        "title": "Title 1",
        "content": "Lorem ipsum...",
      },
      ...
    ]
    í˜•íƒœì˜ ë¦¬ìŠ¤íŠ¸ë¥¼ ë°›ëŠ”ë‹¤.

    center_text: ì¤‘ì•™ì— ë“¤ì–´ê°ˆ í° í…ìŠ¤íŠ¸
    size: SVG ì „ì²´ ì‚¬ì´ì¦ˆ(ì •ì‚¬ê°í˜•)
    outer_ring_radius: ë°”ê¹¥ ì›ë“¤ì´ ìœ„ì¹˜í•  ë°˜ì§€ë¦„(ì¤‘ì‹¬ ê¸°ì¤€)
    item_circle_radius: ë°”ê¹¥ ì›(ì œëª© ì›)ì˜ ë°˜ì§€ë¦„
    """

    # SVG ìƒì„± (ë°˜ì‘í˜• & 1:1 ë¹„ìœ¨)
    dwg = svgwrite.Drawing(
        filename=filename,
        size=("100%", "100%"),
        viewBox=f"0 0 {size} {size}"
    )

    cx = cy = size / 2  # ì¤‘ì•™ ì¢Œí‘œ

    # ----- 1) ì¤‘ì•™ í° ì› -----
    center_circle = dwg.circle(
        center=(cx, cy),
        r=center_circle_radius,
        fill=center_circle_fill
    )
    dwg.add(center_circle)

    # ì¤‘ì•™ ì› ì•ˆì˜ í…ìŠ¤íŠ¸
    dwg.add(dwg.text(
        center_text,
        insert=(cx, cy + 5),  # í…ìŠ¤íŠ¸ ì¤‘ì•™ ì •ë ¬ì„ ìœ„í•œ ë¯¸ì„¸ ì¡°ì •
        fill=center_circle_text_color,
        style=(
            f"font-size:{title_font_size}px;"
            "text-anchor:middle;"
            "dominant-baseline:middle;"  # ìˆ˜ì§ ì¤‘ì•™ ì •ë ¬
        )
    ))

    # ----- 2) ë°”ê¹¥ ë§(í…Œë‘ë¦¬) -----
    # (ì´ë¯¸ì§€ì—ì„œ íšŒìƒ‰ ì› í˜•íƒœë¡œ ë³´ì´ëŠ” ë¶€ë¶„)
    outer_circle = dwg.circle(
        center=(cx, cy),
        r=outer_ring_radius,
        fill="none",
        stroke=outer_ring_stroke,
        stroke_width=outer_ring_stroke_width
    )
    dwg.add(outer_circle)

    # ----- 3) ë°”ê¹¥ ì‘ì€ ì›ë“¤ + í…ìŠ¤íŠ¸ -----
    n = len(data)
    if n == 0:
        dwg.save()
        return

    # ì›ì„ nê°œ ë°°ì¹˜í•  ë•Œ, 360ë„(í˜¹ì€ ì›í•˜ëŠ” ê°ë„)ë¡œ ê· ë“± ë¶„í¬
    # ì—¬ê¸°ì„œëŠ” 360ë„ ì „ì²´ì— 5ê°œ ë°°ì¹˜(ì˜ˆ: ê°ë„ ê°„ê²© = 72ë„)
    # ë§¨ ìœ„(12ì‹œ ë°©í–¥)ë¥¼ ì‹œì‘ì ìœ¼ë¡œ í•˜ê³  ì‹¶ë‹¤ë©´ 270ë„(ìˆ˜í•™ì ìœ¼ë¡œëŠ” -90ë„)ì—ì„œ ì‹œì‘
    # í˜¹ì€ 90ë„ì—ì„œ ì‹œì‘í•´ ì‹œê³„ ë°©í–¥ìœ¼ë¡œ ëŒë ¤ë„ ë¨
    angle_step = 360 / n
    start_angle = -90  # 12ì‹œ ë°©í–¥(ìœ„ìª½)ë¶€í„° ë°°ì¹˜

    for i, item in enumerate(data):
        angle = start_angle + angle_step * i

        # ë°”ê¹¥ ì›(ì œëª© ì›) ì¤‘ì‹¬ ì¢Œí‘œ
        item_cx, item_cy = polar_to_cartesian(cx, cy, outer_ring_radius, angle)

        # ì‘ì€ ì›(ì œëª© í‘œì‹œìš©)
        circle_item = dwg.circle(
            center=(item_cx, item_cy),
            r=item_circle_radius,
            fill=item_circle_fill,
            stroke=item_circle_stroke,
            stroke_width=item_circle_stroke_width
        )
        dwg.add(circle_item)

        # ì‘ì€ ì› ë‚´ë¶€ì— ì œëª© í‘œì‹œ
        dwg.add(dwg.text(
            item["title"],
            insert=(item_cx, item_cy + 4),
            fill=item_circle_stroke,
            style=(
                f"font-size:{title_font_size}px;"
                "text-anchor:middle;"
                "dominant-baseline:middle;"
                "font-weight:bold;"
            )
        ))

        # ----- 4) ë‚´ìš© í…ìŠ¤íŠ¸ ë°°ì¹˜ -----
        # ë‚´ìš© í…ìŠ¤íŠ¸ëŠ” ì‘ì€ ì›ë³´ë‹¤ ì¡°ê¸ˆ ë°”ê¹¥ìª½ì— ë°°ì¹˜
        # (ì›í•˜ëŠ” ë°°ì¹˜ë¥¼ ìœ„í•´ ê±°ë¦¬, ê°ë„, anchor ë“±ì„ ì¡°ì ˆ)
        text_radius = outer_ring_radius + item_circle_radius + 20  # ì‘ì€ ì›ë³´ë‹¤ 20px ë°”ê¹¥
        text_x, text_y = polar_to_cartesian(cx, cy, text_radius, angle)

        # ê°ë„ì— ë”°ë¼ ì •ë ¬ ë°©í–¥(Anchor)ì„ ë°”ê¾¸ê³  ì‹¶ë‹¤ë©´ quadrant ê³„ì‚° ê°€ëŠ¥
        # ì—¬ê¸°ì„œëŠ” ë‹¨ìˆœíˆ ê°€ìš´ë° ì •ë ¬ ì‚¬ìš©
        dwg.add(dwg.text(
            item["content"],
            insert=(text_x, text_y),
            fill="#555555",
            style=(
                f"font-size:{content_font_size}px;"
                "text-anchor:middle;"
            )
        ))

    dwg.save()


if __name__ == "__main__":
    example_data = [
        {
            "title": "Title 1",
            "content": "Lorem ipsum dolor sit amet, simul adolescens ei vis, id nec erram interesset. Ne usu."
        },
        {
            "title": "Title 2",
            "content": "Lorem ipsum dolor sit amet, simul adolescens ei vis, id nec erram interesset. Ne usu."
        },
        {
            "title": "Title 3",
            "content": "Lorem ipsum dolor sit amet, simul adolescens ei vis, id nec erram interesset. Ne usu."
        },
        {
            "title": "Title 4",
            "content": "Lorem ipsum dolor sit amet, simul adolescens ei vis, id nec erram interesset. Ne usu."
        },
        {
            "title": "Title 5",
            "content": "Lorem ipsum dolor sit amet, simul adolescens ei vis, id nec erram interesset. Ne usu."
        },
    ]

    create_circular_diagram(
        data=example_data,
        center_text="TEXT HERE",
        filename="diagram.svg",
        size=600
    )
```
